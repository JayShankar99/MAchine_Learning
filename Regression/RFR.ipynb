# Random Forest Regressor (RFR)

Random Forest Regressor is an Ensemble learning Method.

__Ensemble Learning__:

Ensemble learning is a method that combines the prediction from multiple machine learning algorithms together to make more accurate predictions than any individual model. A model comprised of many models is called an Ensemble model.



Random forest is a Supervised Learning algorithm which uses ensemble learning method for classification and regression.

Random forest is a bagging technique and not a boosting technique. The trees in random forests are run in parallel. There is no interaction between these trees while building the trees.

It operates by constructing a multitude of decision trees at training time and outputting the class that is the mode of the classes (classification) or mean prediction (regression) of the individual trees.

#### Importing Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Loading Dataset

```python
dataset = pd.read_csv("Position_Salaries.csv")
dataset.head()
```

#### Seperating independent and dependent variables

```python
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1]
```

#### Training Random forest model

__Note:__ we dont have to do any feature scaling while using Decission tree and Random forest.

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators=10, random_state=42) #for better results you can tune the perameters.
rfr.fit(X,y)
```

#### Visuaising the decission tree Result High resolution

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X,y,color="blue")
plt.plot(X_grid, rfr.predict(X_grid), color="red")
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
