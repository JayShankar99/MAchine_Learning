# Count or Frequency Encoding

In count or frequency encoding what we do is that we try to replace the categorical feture by the "count" or the "frequency" i.e the number of time the value has occured in a specific column. For example in "Country" column "India" has been repeted for 100 times or we can say the count of "India" is hundred in "Country" column so what we will do we will replace "India" by the count of frequency value of "India" i.e. wherever "India" will be present it will get replaced by 100.

The disadvantage of this encoding method is that it will note be able to handle those situation when two or more categories will have same count or frequency.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("adult.csv")
df.head()
```

```python
df.info()
```

```python
for i in range(1,16):
    if df[str(i)].dtype!="int64":
        print("Column {} {} \n".format(i,df[str(i)].unique()))
```

```python
for i in range(1,16):
    if df[str(i)].dtype!="int64":
       df[str(i)].replace(" ?","Unknown",inplace=True)
```

```python
column = []
for i in range(1,15):
    if df[str(i)].dtype!="int64":
        column.append(str(i))
column
```

```python
df = df[column]
df.columns=["Employe","Degree","Status","Post","Family_status","Races","Gender","Country"]
df.head()
```

```python
for column in df.columns[:]:
    print("{} : {} values".format(column,len(df[column].unique())))
```

## Count or Frequency encoding

```python
df["Country"].unique()
```

```python
df["Country"].value_counts()
```

```python
dictt = df["Country"].value_counts().to_dict()
dictt
```

```python
df["Country"] = df["Country"].map(dictt)
df.head()
```
